```python
#Q-2--&> Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?

import numpy as np
import matplotlib.pyplot as plt

values = np.arange(1, 7)
probabilities = np.ones(len(values)) / len(values)  # Equal probability for each value

# Display the PMF
for v, p in zip(values, probabilities):
    print(f"Value: {v}, Probability: {p}")

# Plot the PMF
plt.stem(values, probabilities)
plt.title("Discrete Uniform Distribution PMF")
plt.xlabel("Values")
plt.ylabel("Probability")
plt.ylim(0, 1)
plt.show()

```

```python
#Q-3- Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?
def bernoulli_pmf(p, x):
    """

    Parameters:
    p : float
        Probability of success (0 <= p <= 1)
    x : int
        Outcome (0 or 1)

    Returns:
    float : Probability of outcome x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for Bernoulli distribution")
    return p**x * (1 - p)**(1 - x)

# Example usage
p = 0.7  # Probability of success
for x in [0, 1]:
    print(f"P(X={x}) = {bernoulli_pmf(p, x)}")

```

```python
#Q-6--> Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
#uniform distribution.


# Define discrete uniform distribution (values 1 to 6)
values = np.arange(1, 7)
probabilities = np.ones(len(values)) / len(values)  # equal probabilities

# Calculate CDF
cdf = np.cumsum(probabilities)

# Display CDF
for v, c in zip(values, cdf):
    print(f"Value: {v}, CDF: {c}")

# Plot CDF
plt.step(values, cdf, where='mid', marker='o')
plt.title("CDF of Discrete Uniform Distribution")
plt.xlabel("Values")
plt.ylabel("Cumulative Probability")
plt.ylim(0, 1.05)
plt.grid(True)
plt.show()
```

```python
#Q-9--Write a Python function to calculate Z-scores from a dataset and plot them?

def calculate_and_plot_zscores(data):

    data = np.array(data)
    
    mean = np.mean(data)
    std = np.std(data)
    
    # Z-scores
    z_scores = (data - mean) / std
    
    for i, z in enumerate(z_scores):
        print(f"Data point {data[i]} -> Z-score: {z:.2f}")
    
    # Plot Z-scores
    plt.figure(figsize=(8,4))
    plt.stem(data, z_scores)
    plt.title("Z-scores of Dataset")
    plt.xlabel("Data Values")
    plt.ylabel("Z-score")
    plt.grid(True)
    plt.show()
    
    return z_scores

# Example usage
dataset = [10, 12, 15, 18, 20, 25]
z_scores = calculate_and_plot_zscores(dataset)

```

```python
#Q--Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)?

from scipy.stats import norm

def plot_standard_normal():
  
    x = np.linspace(-4, 4, 1000)  # 4 std deviations on each side
    
    pdf = norm.pdf(x, loc=0, scale=1)
    
    # Plot the standard normal distribution
    plt.plot(x, pdf, color='blue')
    plt.fill_between(x, pdf, alpha=0.2, color='blue')
    plt.title("Standard Normal Distribution (μ=0, σ=1)")
    plt.xlabel("Value")
    plt.ylabel("Probability Density")
    plt.grid(True)
    plt.show()

# Example usage
plot_standard_normal()

```

```python
#Q-Implement hypothesis testing using Z-statistics for a sample dataset?

sample = [52, 55, 50, 53, 54, 56, 52, 51, 57, 53]

mu_0 = 50     
sigma = 3     

n = len(sample)
sample_mean = np.mean(sample)

z = (sample_mean - mu_0) / (sigma / np.sqrt(n))
print(f"Sample mean: {sample_mean:.2f}")
print(f"Z-statistic: {z:.2f}")

# Significance level
alpha = 0.05

# Two-tailed test critical value
z_critical = norm.ppf(1 - alpha/2)
print(f"Critical Z-value for α={alpha}: ±{z_critical:.2f}")

# Decision
if abs(z) > z_critical:
    print("Reject the null hypothesis (H0)")
else:
    print("Fail to reject the null hypothesis (H0)")

```

```python
#Q-Create a confidence interval for a dataset using Python and interpret the result.

data = [52, 55, 50, 53, 54, 56, 52, 51, 57, 53]

# Sample statistics
n = len(data)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation

# Confidence level
confidence = 0.95
alpha = 1 - confidence

# Z-critical value for 95% confidence
z_critical = norm.ppf(1 - alpha/2)

# Margin of error
margin_of_error = z_critical * (sample_std / np.sqrt(n))

# Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Interpretation
print(f"We are 95% confident that the true population mean lies between {ci_lower:.2f} and {ci_upper:.2f}.")

```

```python
#Q-Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

lambda_ = 3  
x_values = np.arange(0, 11)  

#  CDF
cdf_values = poisson.cdf(x_values, lambda_)

for x, cdf in zip(x_values, cdf_values):
    print(f"P(X ≤ {x}) = {cdf:.4f}")

# Plot CDF
plt.step(x_values, cdf_values, where='mid', marker='o')
plt.title(f"Cumulative Distribution Function (Poisson, λ={lambda_})")
plt.xlabel("Number of Occurrences")
plt.ylabel("Cumulative Probability")
plt.ylim(0, 1.05)
plt.grid(True)
plt.show()

# Interpretation
print(f"Interpretation: For example, P(X ≤ 4) = {poisson.cdf(4, lambda_):.4f} means there is a "
      f"{poisson.cdf(4, lambda_)*100:.2f}% chance that the number of occurrences is 4 or less.")


```

```python
#Q-Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution?

mu = 50      # mean
sigma = 10   # standard deviation
size = 100   # number of samples

data = np.random.normal(mu, sigma, size)

data_range = np.max(data) - np.min(data)

q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Display results
print(f"Generated dataset (first 10 values): {data[:10]}")
print(f"Range of the dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")

```

```python
#Q-Implement Z-score normalization on a dataset and visualize its transformation5.
import numpy as np
import matplotlib.pyplot as plt

# Example dataset
data = np.array([10, 12, 15, 18, 20, 25, 22, 17])

# Calculate mean and standard deviation
mean = np.mean(data)
std = np.std(data)

# Z-score normalization
z_scores = (data - mean) / std

# Display results
print("Original data:", data)
print("Z-score normalized data:", np.round(z_scores, 2))

# Visualization
plt.figure(figsize=(10,5))
plt.plot(data, label='Original Data', marker='o')
plt.plot(z_scores, label='Z-score Normalized', marker='o')
plt.title("Z-score Normalization of Dataset")
plt.xlabel("Index")
plt.ylabel("Value / Z-score")
plt.legend()
plt.grid(True)
plt.show()

```
