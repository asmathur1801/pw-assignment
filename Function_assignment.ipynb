Q-1 >> What is the difference between a function and a method in Python?

Ans-Function: A standalone block of code defined using def, not tied to any object.

Method: A function defined inside a class; it works on objects of that class and can access/modify object state.



````python
#Example:
def add(a, b):       # function

    return a + b

class Calculator:

    def add(self, a, b):   # method
        return a + b

print(add(2,3))                 # 5

print(Calculator().add(2,3))    # 5

````

Q4. What is the purpose of the return statement in a Python function?

It sends a value back to the caller and ends function execution.

A function without return gives None.

````python
#Example:
def divide(a, b):
    if b == 0:
        return None
    return a/b

print(divide(10,2))  # 5.0
````

Q5. What are iterators in Python and how do they differ from iterables?

Iterable: An object you can loop over (list, tuple, string, etc.), it has __iter__().

Iterator: An object with __next__() that gives items one by one and raises StopIteration when done.


````python
#Example:
lst = [1,2,3]
it = iter(lst)    # iterator
print(next(it))   # 1
print(next(it))   # 2
````

Q6. Explain the concept of generators in Python and how they are defined.

A generator is a special iterator created using a function with the yield keyword.

They produce values lazily (one at a time).

````python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

for x in count_up_to(3):
    print(x)   # 1 2 3
````

Q8. What is a lambda function in Python and when is it typically used?

A lambda function is a small, anonymous function written in one line: lambda args: expression.

Used for short, simple tasks (like sorting keys, map, filter).

````python
pairs = [(1,3), (2,1), (3,2)]
pairs.sort(key=lambda t: t[1])
print(pairs)  # [(2,1), (3,2), (1,3)]
````

Q10. What is the difference between map(), reduce(), and filter()?

map: transforms every element.

filter: selects elements that meet a condition.

reduce: combines all elements into a single result.

````python
from functools import reduce
nums = [1,2,3,4]
print(list(map(lambda x: x*2, nums)))         # [2,4,6,8]
print(list(filter(lambda x: x%2==0, nums)))   # [2,4]
print(reduce(lambda a,b: a+b, nums))          # 10
````

````python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1: return False
    if n <= 3: return True
    if n % 2 == 0: return False
    i = 3
    while i*i <= n:
        if n % i == 0:
            return False
        i += 2
    return True

primes = [n for n in range(1,201) if is_prime(n)]
print(primes)
````

````python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
#terms.
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n = n_terms
        self.index = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= self.n:
            raise StopIteration
        if self.index == 0:
            self.index += 1
            return 0
        elif self.index == 1:
            self.index += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.index += 1
            return self.b

for x in FibonacciIterator(10):
    print(x, end=" ")
````

````python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent
def powers_of_two(max_exponent):
    for e in range(max_exponent+1):
        yield 2 ** e

print(list(powers_of_two(5)))  # [1,2,4,8,16,32]
````

````python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_lines_generator(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            yield line.rstrip('\n')

# Example: read from sample file
with open("sample.txt","w") as f:
    f.write("Line1\nLine2\nLine3\n")

for line in read_lines_generator("sample.txt"):
    print(line)

````

````python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit
def c_to_f(c):
    return c * 9/5 + 32

celsius = [0, 20, 37, 100]
fahrenheit = list(map(c_to_f, celsius))
print(fahrenheit)  # [32.0, 68.0, 98.6, 212.0]
````

````python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(s):
    vowels = set("aeiouAEIOU")
    return ''.join(filter(lambda ch: ch not in vowels, s))

print(remove_vowels("Beautiful Day"))  # Btfl Dy
````
