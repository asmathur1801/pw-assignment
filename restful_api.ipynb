Question 2: What is Flask, and why is it popular for building APIs?

ANS->Flask is a Python-based web framework that helps developers create web applications and APIs in a simple and flexible way. It is called a lightweight framework because it provides only the basic features needed to handle web requests, without adding unnecessary complexity.

Flask is popular for building APIs because it is easy to learn and allows developers to start building APIs quickly. It does not force a fixed structure, so developers have the freedom to design their application according to their needs. Flask also supports all common HTTP methods like GET, POST, PUT, and DELETE, which are essential for RESTful APIs and also works well with json .

Question 3:What are HTTP methods used in RESTful APIs?

ANS-->HTTP methods in RESTful APIs are used to define what action the client wants to perform on a resource (data stored on the server).

The main HTTP methods used in RESTful APIs are:

GET – Used to retrieve data from the server. It does not change the data.

POST – Used to send new data to the server and create a new resource.

PUT – Used to update an existing resource completely.

PATCH – Used to partially update an existing resource.

DELETE – Used to remove a resource from the server.

Question 4: What is the purpose of the @app.route() decorator in Flask?

ANS->The @app.route() decorator in Flask is used to connect a URL with a specific function in the application. It tells Flask which function should run when a user visits a particular URL.

Question 5: What is the role of Flask-SQLAlchemy?

ANS-->Flask-SQLAlchemy is an extension of Flask that is used to connect a Flask application with a database and manage data easily. It provides an Object Relational Mapping (ORM) system, which allows developers to work with database tables using Python classes instead of writing complex SQL queries.

```python
# Question- 6: How do you create a basic Flask application?

from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return "Hello, Flask!"

if __name__ == '__main__':
    app.run()
```

```python
#Question 7:How do you return JSON responses in Flask?

#ANS-->In Flask, JSON responses are returned to send structured data from the server to the client,
#especially in APIs. Flask provides a built-in function called jsonify() to return data in JSON format easily

from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/data')
def data():
    return jsonify({"message": "Hello", "status": "success"})
```

```python
#Question 8:How do you handle POST requests in Flask?

#ANS->In Flask, POST requests are used when the client wants to send data to the server, like submitting a form or creating a new resource.
#To handle POST requests, we need to specify the HTTP method in the @app.route() decorator and then use request to access the sent data.
from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route('/submit', methods=['POST'])
def submit():
    data = request.json  # Get JSON data sent by client
    return jsonify({"received_data": data})
```

```python
# Question 9: How do you handle errors in Flask (e.g., 404)?
# ANS--> In Flask, errors like 404 Not Found can be handled using error handlers.
#This allows the application to return a custom message or page instead of the default error page.
from flask import Flask, jsonify

app = Flask(__name__)

@app.errorhandler(404)
def not_found(e):
    return jsonify({"error": "Page not found"}), 404
```

```python
# Question 10: How do you structure a Flask app using Blueprints?
#In Flask, Blueprints are used to organize a large application into smaller, reusable modules. Instead of writing all routes and logic in a single file,
#Blueprints let you split your app by features, making it easier to maintain and scale.

# example:-  Create a Blueprint in a separate file for a specific feature:
from flask import Blueprint

user_bp = Blueprint('user', __name__)

@user_bp.route('/profile')
def profile():
    return "User Profile"

# Register the Blueprint in the main app file:
from flask import Flask
from user import user_bp

app = Flask(__name__)
app.register_blueprint(user_bp, url_prefix='/user')
```
