Q-1>>What are data structures, and why are they important?

Ans>>>>A data structure is a way of organizing, storing, and managing data so that it can be used efficiently.

Q_2>>> Explain the difference between mutable and immutable data types with examples?

Ans->>>A mutable data type is a type whose value can be changed or modified after the object is created, without creating a new object in memory.
 Example: List, Dictionary, Set


 An immutable data type is a type whose value cannot be changed after the object is created. If you try to modify it, a new object is created in memory.
Example: String, Tuple, int, float

Q-3>>>>What are the main differences between lists and tuples in Python?

ANS>>>>A list in Python is a mutable, ordered collection of elements that can store different data types.

A tuple in Python is an immutable, ordered collection of elements that can also store different data types.

Q-5>>>>Why might you use a set instead of a list in Python?

Ans>>>A set in Python is an unordered collection of unique elements. We use a set instead of a list when we want to avoid duplicates, perform faster membership testing, and carry out mathematical operations like union, intersection, and difference. Unlike lists, sets automatically remove duplicate values and provide faster lookups.

Q-17>>>Describe the time complexity of accessing elements in a dictionary.

Ans->>Average case: O(1) (constant time) using hashing.

Worst case: O(n) (if many keys hash to the same value, causing collisions).
