Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with
examples

ANS--->Descriptive statistics: summarize and describe the features of a dataset you have.
Examples:
mean, median, mode, standard deviation, frequency tables, histograms.
Example: reporting the average marks of 100 students = descriptive.
---> Inferential statistics: use sample data to make conclusions or predictions about a larger population,and quantify uncertainty (confidence intervals, hypothesis tests, regression).

Example: using a sample of 400 voters to estimate the support for a candidate in the whole city, with a
margin of error

Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.

ANS-->Sampling: selecting a subset (sample) from a population to measure and draw conclusions about the population.

-->Random sampling: every unit in the population has an equal chance to be selected (e.g.,drawing random phone numbers).
 Good for unbiased representation when population is homogeneous.

--> Stratified sampling: divide population into meaningful groups (strata) —

 e.g., age groups or regions — then sample randomly within each stratum. Use when important subgroups must be represented proportionally (reduces variance and ensures representation of smaller groups).


Question 3: Define mean, median, and mode. Explain why these measures of central tendency
are important.

ANS-->>Mean: arithmetic average (sum of values ÷ count). Good for further calculations and
when distribution is symmetric.

-->Median: middle value when data are ordered (or average of two middle values). Robust to
outliers; better center measure when data skewed.

->> Mode: most frequent value(s). Useful for categorical data and to detect multimodality.
 These measures are important because they describe the ―center or typical value of data and
guide decisions, modeling, and interpretation.

: Implement a Python program to compute the mean, median, and mode of a given list of
numbers. numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


```python
#Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers.
#numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
from collections import Counter

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Mean
mean = sum(numbers) / len(numbers)

# Median
numbers_sorted = sorted(numbers)
n = len(numbers_sorted)

if n % 2 == 0:
    median = (numbers_sorted[n//2 - 1] + numbers_sorted[n//2]) / 2
else:
    median = numbers_sorted[n//2]

# Mode
count = Counter(numbers)
max_freq = max(count.values())
mode = [num for num, freq in count.items() if freq == max_freq]

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
#Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python
#list_x = [10, 20, 30, 40, 50]
#list_y = [15, 25, 35, 45, 60]
import math

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

n = len(list_x)

# Mean of X and Y
mean_x = sum(list_x) / n
mean_y = sum(list_y) / n

# Covariance
covariance = sum((list_x[i] - mean_x) * (list_y[i] - mean_y) for i in range(n)) / (n - 1)

# Standard deviation of X and Y
std_x = math.sqrt(sum((x - mean_x) ** 2 for x in list_x) / (n - 1))
std_y = math.sqrt(sum((y - mean_y) ** 2 for y in list_y) / (n - 1))

# Correlation coefficient
correlation = covariance / (std_x * std_y)

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)

```

```python
#Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
#data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Draw boxplot
plt.boxplot(data)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

# Identify outliers using IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```
